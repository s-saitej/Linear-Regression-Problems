```python
import pandas as pd
```

```python
dataframe = pd.read_csv('concrete.csv')
```

```python
dataframe
```

```python
dataframe.isnull()

dataframe.shape

x = dataframe.iloc[:515,:]
y = dataframe.iloc[515:,:]
```

```python
# Data Selection

from sklearn import model_selection as ms
```

```python
first_data,second_data,first_target,seconda_target = ms.train_test_split(x,y,random_state=3)
```

```python
# Running regression procedure

from sklearn import linear_model as lm
```

```python
regression = lm.LinearRegression()
```

```python
# Fitting a curve

fitting = regression.fit(first_data,first_target)
```

```python
# Predicting a value

result = fitting.predict(first_data)
result
```

```python
# To know the Co-efficients

print("Coefficients are "+str(regression.coef_))
```

```python
# To know the intercepts

print("Intercepts are"+str(regression.intercept_))
```

```python
# To find errors 

from sklearn import metrics as met


print("Mean Squared Error is "+str(met.mean_squared_error(first_data,result)))
```

```python
# To find varience

print("Varience is "+str(met.r2_score(first_data,result)))
```
